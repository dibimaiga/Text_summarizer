```python
from dataclasses import dataclass
from pathlib import Path

@dataclass
class DataValidationConfig:
    root_dir: Path
    status_file: str
    all_required_files: list
```

```python
from TextSummarizationProject.constants import *
from TextSummarizationProject.utils.common import read_yaml, create_directories
```

```python
class ConfigurationManager:
    def __init__(self, 
                 config_filepath: Path = CONFIG_FILE_PATH, 
                 params_filepath: Path = PARAMS_FILE_PATH):
        self.config = read_yaml(config_filepath)
        self.params = read_yaml(params_filepath)

        create_directories([self.config.artefacts_root])

    
    def get_data_validation_config(self) -> DataValidationConfig:
        try:
            config = self.config.data_validation

            create_directories([config.root_dir])

            data_validation_config = DataValidationConfig(
                root_dir=config.root_dir,
                status_file=config.status_file,
                all_required_files=config.all_required_files
            )

            return data_validation_config
        
        except Exception as e:
            raise e
```

```python
import os
import zipfile
from TextSummarizationProject.utils.common import get_size
import urllib.request as request
from TextSummarizationProject.logging import logger
```

```python
class DataValidation:
    def __init__(self, config: DataValidationConfig):
        self.config = config
    
    def validate_all_files(self) -> bool:
        try:
            status = None

            all_files = os.listdir(os.path.join("artefacts", "data_ingestion","samsum_dataset"))

            for file in all_files:
                if file not in self.config.all_required_files:
                    status = False
                    with open(self.config.status_file, "w") as f:
                        f.write(f"Validation status: {status}")

                else:
                    status = True
                    
                    with open(self.config.status_file, "w") as f:     
                        f.write(f"Validation status: {status}")
                        
            return status
        
        except Exception as e:
            raise e
```

```python
try:
    config = ConfigurationManager()
    data_validation_config = config.get_data_validation_config()
    data_validation = DataValidation(config=data_validation_config)
    data_validation.validate_all_files()
except Exception as e:
    raise e
```
